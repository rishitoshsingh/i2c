```python
import torch
from torch.utils.data import DataLoader
from datasets import Flickr30kDataset
import matplotlib.pyplot as plt
```

```python
import torchvision.transforms as transforms

# Define the transform function
transform = transforms.Compose([
    # transforms.Resize((256, 256)),
    transforms.ToTensor()
])
```

```python
dataset = Flickr30kDataset("/Volumes/T7 Shield/Datasets/Flickr30", transform=transform)
dataloader = DataLoader(dataset, batch_size=1, shuffle=True)
```

```python
def plot_images_with_captions(images, captions):
    fig, axes = plt.subplots(nrows=len(images), ncols=2, figsize=(7, 7))
    for i, (image, caption) in enumerate(zip(images, captions)):
        axes[i, 0].imshow(image.permute(1, 2, 0))
        axes[i, 0].axis('off')
        axes[i, 1].text(0, 0.5, caption, wrap=True, ha='left', va='center', fontsize=10)
        axes[i, 1].axis('off')
        # Set the x-axis limit to the image width
        axes[i, 1].set_xlim([0, image.shape[2]])
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
from textwrap import wrap

# Get the first 5 images and captions from the DataLoader
images = []
captions = []
for i, (image, caption) in enumerate(dataloader):
    images.append(image[0])
    captions.append(caption[0])
    if i == 4:
        break

# Plot the images with captions
plot_images_with_captions(images, captions)
```
